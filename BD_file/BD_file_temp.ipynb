```python
import pandas as pd
```

## 시군구 정리 코드
데이터 출처 : https://blog.naver.com/s9soft/221530595045

```python
df = pd.read_csv('./ETC/시도_시군구_20190531.csv',encoding='cp949',header=None)

df.columns

df['시도_시군구'] = df.apply(lambda x : str(x[0]) + ' ' + str(x[1]).split()[0],axis=1)


df = df.drop_duplicates(subset=['시도_시군구'])

df['시도_시군구'].to_csv('./ETC/시군구_최종정리.csv')
```
